## accuracy = 2

```python
import matplotlib.pyplot as plt
import numpy as np
import os
os.chdir("/Users/ahasan/C++/")

data = np.loadtxt('seis2.txt')
x = data[:, 0]
y = data[:, 1]

plt.figure(figsize=(12, 4))
plt.plot(x, y, label='Seismogram', color='blue')
plt.xlabel('Time (s)', fontsize=12)
plt.ylabel('Amplitude', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```
